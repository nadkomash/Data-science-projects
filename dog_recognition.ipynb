```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from skimage.io import imread
from skimage.transform import resize
import matplotlib.pyplot as plt
```

```python
model = ResNet50(weights='imagenet')
from skimage.io import imread
pluto = imread('/content/pluto.jpg')
fig , ax = plt.subplots(figsize=(7,4))
_=ax.imshow(pluto)
```

```python
pluto = resize(pluto, (224,224))
pluto = pluto.reshape(1,224,224,3)*255
pluto = preprocess_input(pluto)
pluto_pred = model.predict(pluto,verbose=0)
print(pluto_pred.shape)
```

```python
pluto_breed = decode_predictions(pluto_pred,top=5)

for _, breed,score in pluto_breed[0]:
  print('%s: %.2f' % (breed, score))
```
